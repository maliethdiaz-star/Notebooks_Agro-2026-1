### 3 Ejemplos Sencillos para Agroindustria

Aquí tienes tres fragmentos de código básicos que ilustran tareas comunes:

#### 1. Cálculo del Índice de Madurez (Relación Sólidos/Acidez)

Este cálculo es vital para determinar el momento óptimo de cosecha en frutas como los cítricos.

```
# Datos de entrada
grados_brix = 12.5
acidez_titulable = 0.85

# Cálculo del índice
indice_madurez = grados_brix / acidez_titulable

print(f"El índice de madurez es: {indice_madurez:.2f}")

if indice_madurez > 12:
    print("Estado: Apto para cosecha/procesamiento.")
else:
    print("Estado: Aún no alcanza la madurez óptima.")
```

#### 2. Estimación de Capacidad de Almacenamiento

Calcula cuántas toneladas de grano caben en un silo cilíndrico estándar.

```
import math

radio = 3.5 # metros
altura = 10 # metros
densidad_maiz = 0.75 # toneladas por metro cúbico

# Fórmula del volumen del cilindro: V = pi * r^2 * h
volumen = math.pi * (radio ** 2) * altura
capacidad_toneladas = volumen * densidad_maiz

print(f"La capacidad estimada del silo es de {capacidad_toneladas:.2f} toneladas.")
```

#### 3. Conversión de Unidades de Temperatura (Pasteurización)

Un conversor simple para procesos térmicos donde se manejan diferentes escalas.

```
def celsius_a_fahrenheit(c):
    return (c * 9/5) + 32

temp_pasteurizacion = 72 # Grados Celsius

print(f"Para la pasteurización HTST, se requieren {celsius_a_fahrenheit(temp_pasteurizacion)} °F.")
```
